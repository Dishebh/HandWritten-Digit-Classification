#### def dist (a1, a2):
    return np.sum((a1 - a2)**2)**.5

```python
def KNN (X, Y, test_point, k = 5):
    
    m = X.shape[0]
    vals = []
    
    for i in range(m):
        d = dist (X[i], test_point)
        vals.append((d, Y[i]))
        
    vals = sorted (vals)
    vals = np.array (vals[:k])
    b = np.unique (vals[:, 1], return_counts = True)
    idx = np.argmax (b[1])
    
    pred = b[0][idx]
    
    return (int(pred))
```

```python
KNN (X, Y, test_fruit)
```
